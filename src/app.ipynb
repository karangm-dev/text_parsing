```python
import glob
import re
import spacy
## Number of characters 5175387
nlp = spacy.load('en_core_web_sm') 
import heapq
```

```python
output_file = open('../output/Q1.txt', 'w')
```

```python
def load_data(input_directory):
    file_contents_list = []
    for file_name in glob.glob(input_directory + "/*"):
        print("Prepocessing: {}".format(file_name))
        file_pointer = open(file_name, "r")

        # Read file contents
        file_content = file_pointer.read()

        # Remove duplicate spaces
        file_content = re.sub(' +', ' ', file_content)

        # Remove new line characters
        file_content = file_content.replace("\n", " ")

        file_contents_list.append(file_content)   
        file_pointer.close()
    return file_contents_list
```

```python
def get_parsed_by_spacy(file_contents_list):
    doc_list = []
    for text in file_contents_list:
        doc = nlp(text)
        doc_list.append(doc)
    return doc_list
```

```python
def run_1():
    print("##### Loading data")
    data = load_data('../input/dataset')
#     doc = get_parsed_by_spacy(data[:1010])
    doc_list = get_parsed_by_spacy(data)
    return doc_list
```

```python
doc_list = run_1()
```

### 1.1 [2 points]

Report the number of sentences parsed; do so by searching for “ROOT” in the dependency representation or use the sentence segmentation method of spacy parser.

```python
def get_number_of_sentences(doc_list):
    sentence_count = 0
    for doc in doc_list:
        for token in doc:
            if token.dep_ == 'ROOT':
                sentence_count = sentence_count + 1
    print('Sentence Count: {} \n'.format(sentence_count))
    output_file.write('Sentence Count: {} \n'.format(sentence_count))
    return sentence_count
```

```python
output_file.write('\n ####### Q1.1 \n')
sentence_count = get_number_of_sentences(doc_list)
```

### 1.2 [2 points]
Using the part-of-speech tags that the parser has given you, report the average number of verbs per sentence in the overall corpus (i.e. total number of verbs divided by total number of sentences in the corpus). Use the total number of sentences from 1.1. Also report the part-of-speech tags that you are using to identify the verbs.

```python
def get_number_of_verbs(doc_list, sentence_count):
    vb_count = 0
    vbd_count = 0
    vbg_count = 0
    vbn_count = 0
    vbp_count = 0
    vbz_count = 0
    
    for doc in doc_list:
        for token in doc:
            if token.tag_ == 'VB':
                vb_count = vb_count + 1
            elif token.tag_ == 'VBD':
                vbd_count = vbd_count + 1
            elif token.tag_ == 'VBG':
                vbg_count = vbg_count + 1
            elif token.tag_ == 'VBN':
                vbn_count = vbn_count + 1
            elif token.tag_ == 'VBP':
                vbp_count = vbp_count + 1
            elif token.tag_ == 'VBZ':
                vbz_count = vbz_count + 1
            
    print("VB: {} \nVBD: {} \nVBG: {} \nVBN: {} \nVBP: {} \nVBZ: {}".format(
        vb_count, vbd_count, vbg_count, vbn_count, vbp_count, vbz_count))
    
    output_file.write("VB: {} \nVBD: {} \nVBG: {} \nVBN: {} \nVBP: {} \nVBZ: {}\n".format(
        vb_count, vbd_count, vbg_count, vbn_count, vbp_count, vbz_count))
    
    print("Average --- \nVB: {} \nVBD: {} \nVBG: {} \nVBN: {} \nVBP: {} \nVBZ: {}".format(
        vb_count/sentence_count, vbd_count/sentence_count, vbg_count/sentence_count,
        vbn_count/sentence_count, vbp_count/sentence_count, vbz_count/sentence_count))
    
    output_file.write("Average --- \nVB: {} \nVBD: {} \nVBG: {} \nVBN: {} \nVBP: {} \nVBZ: {}\n".format(
        vb_count/sentence_count, vbd_count/sentence_count, vbg_count/sentence_count,
        vbn_count/sentence_count, vbp_count/sentence_count, vbz_count/sentence_count))

    total = vb_count + vbd_count + vbg_count + vbn_count + vbp_count + vbz_count
    average = total/sentence_count
    
    print("Total average verbs per sentance: {}".format(average))
    output_file.write("Total average verbs per sentance: {}".format(average))
```

```python
output_file.write('\n ####### Q1.2 \n')
get_number_of_verbs(doc_list, sentence_count)
```

### 1.3 [3 points]
Using the dependency tree representation that the parser has given you, report the total number of prepositions found in the overall corpus. In addition, report top three most commonly used preposition.

```python
def get_number_of_prepositions(doc):
    prepositions_count = 0
    prepositions_dict = {}
    
    for doc in doc_list:
        for token in doc:
            if token.tag_ == 'IN':
                prepositions_count = prepositions_count + 1
                if token.text in prepositions_dict:
                    prepositions_dict[token.text ] = prepositions_dict[token.text ] + 1
                else:
                    prepositions_dict[token.text ] = 1
    
    print("Number of prepositions: {}\n".format(prepositions_count))
    output_file.write("Number of prepositions: {}\n".format(prepositions_count))

    ## Find top three prepositions
    prepostion_count_list = list(prepositions_dict.items())
    prepostion_count_list = [t[::-1] for t in prepostion_count_list]    
    heapq.heapify(prepostion_count_list)
    top_three_prepositions = heapq.nlargest(3, prepostion_count_list)
    print("Top three prepositions: {}\n".format(top_three_prepositions))
    output_file.write("Top three prepositions: {}\n".format(top_three_prepositions))
    
```

```python
output_file.write('\n ####### Q1.3 \n')
get_number_of_prepositions(doc_list)
```

### 1.4 [3 points]
Using the named entity recognition method of the parser, report the total number of entities found in the overall corpus. Also, list out all the unique entity labels found in the corpus (e.g. PERSON, ORG etc.).

```python
def get_number_of_entities(doc_list):
    entities_count = 0
    entities_dict = {}
    
    for doc in doc_list:
        for ent in doc.ents:
            entities_count = entities_count + 1
            if ent.label_ in entities_dict:
                entities_dict[ent.label_] = entities_dict[ent.label_] + 1
            else:
                entities_dict[ent.label_] = 1
     
    print("Entities count: {}\n".format(entities_count))   
    output_file.write("Entities count: {}\n".format(entities_count))
    
    print("Entities in corpus: {}\n".format(entities_dict.keys()))   
    output_file.write("Entities in corpus: {}\n".format(entities_dict.keys()))
```

```python
output_file.write('\n ####### Q1.4 \n')
get_number_of_entities(doc_list)
```
